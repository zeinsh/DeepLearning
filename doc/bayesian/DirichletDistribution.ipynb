```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def plotDD(alpha):
    CONST=2/np.sqrt(3)
    xa=[]
    ya=[]
    for i in range(0,1000):
        s=np.random.dirichlet(alpha)
        x=s[0]
        y=s[1]
        z=s[2]
        xa.append(CONST*(y-x+1)/2)
        ya.append(CONST*z)
    plt.figure(figsize=(5,4))
    plt.xlim((-0.01,CONST+0.01))
    plt.ylim((-0.01,CONST+0.01))
    plt.text(0,0,'x')
    plt.text(CONST/2,CONST,'y')
    plt.text(CONST,0,'z')
    plt.scatter(xa,ya)
    plt.plot([0,CONST/2],[0,CONST])
    plt.plot([0,CONST/2],[0,CONST])
    plt.plot([0,CONST],[0,0])
    plt.plot([CONST/2,CONST],[CONST,0])
```

```python
alpha=[1,1,1]
plotDD(alpha)
```

```python
alpha=[5,2,2]
plotDD(alpha)
```

```python
alpha=[5,5,2]
plotDD(alpha)
```

```python
alpha=[0.1,0.1,0.1]
plotDD(alpha)
```

```python
from scipy.special import gamma
def beta(alpha):
    nom=1./gamma(np.sum(alpha))
    for a in alpha:
        nom=nom*gamma(a)
    return nom
def f(x,alpha):
    ret=1
    for i in range(len(x)):
        ret*=np.power(x[i],alpha[i])
    return ret/beta(alpha)
alpha=[.1,0.1,0.1]
x=[0.1,0,0.99]
f(x,alpha)
```
