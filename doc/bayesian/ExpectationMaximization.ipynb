## KL divergence

```python
from lib.bayesian1Dmd import *
import numpy as np
```

```python
def KL(q,p,scaleF):
    return np.sum(np.multiply(q,np.log(q/p)))*scaleF
```

```python
#p(t|x,theta)
dist1=Normal(-500,500,0.001,M=0,S=100)
dist2=Normal(-500,500,0.001,M=1,S=100)
#dist2.getDist()
q=dist2.getDist()
p=dist1.getDist()
print("KL(q||q)={}\nKL(q||p)={}".format(KL(q,q,0.001),KL(q,p,0.001)))

plt.figure()
dist1.plot()
dist2.plot()
plt.grid()
plt.legend(["N(0,100)","N(1,100)"])
```
