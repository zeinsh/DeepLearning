```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
a1=np.arange(50)*3+np.random.randint(-20,20,size=50)
a2=np.arange(50)*3.6+np.random.randint(-15,15,size=50)
plt.plot(a1)
plt.plot(a2)
plt.grid()
```

```python
print(np.cov(a1,a2))
```

```python
print(a1.var(),a2.var())
```

### Sample from multi-variate Gaussian Distribution

```python
a=np.random.randn(100)*50
a1,a2=a[:50],a[50:]
plt.plot(a1)
plt.plot(a2)
plt.grid()
```

```python
print(np.cov(a1,a2))
```
